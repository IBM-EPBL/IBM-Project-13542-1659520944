## **Replace Outliers**

```python
qnt=dt.quantile(q=(0.25,0.75))
tmp = qnt.loc[0.75]-qnt.loc[0.25]
low = qnt.loc[0.25] - 1.5*tmp
high = qnt.loc[0.75] + 1.5 * tmp
dt["Age"]=np.where(dt["Age"]<25,50,dt["Age"])
sns.boxplot(dt["Age"])
```

# **Check for Categorical columns and perform encoding.**

```python
from sklearn.preprocessing import LabelEncoder
dt.head(5)
```

```python
enc=LabelEncoder()
encode1=enc.fit_transform(np.array(dt['Geography']))
encode2=enc.fit_transform(np.array(dt['Gender']))
encode3=enc.fit_transform(np.array(dt['Surname']))
encode1
```

```python
dt['Geography']=encode1
dt['Gender']=encode2
dt['Surname']=encode3
dt.head(5)
```

# **Split the data into dependent and independent variables.**

```python
y=dt['Exited']
print(y)
x=dt.drop('Exited',axis=1)
x
```

# **Scale the independent variables**

```python
from sklearn.preprocessing import scale
y=scale(y)
y
```

# **Split the data into training and testing**

```python
x=np.array(x)
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=40)
```
